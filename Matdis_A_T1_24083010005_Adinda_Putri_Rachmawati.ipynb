```python
import pandas as pd
df = pd.read_csv('data mahasiswa matdis.csv')
df
```

```python
#Tentukan bobot matriks P
import numpy as np
P = np.array([0.3, 0.7, 0.6, 0.8])

#Replace comma with dot in 'ASYIK DAN MENARIK' column and convert to numeric
df['ASYIK DAN MENARIK'] = df['ASYIK DAN MENARIK'].astype(str).str.replace(',', '.').astype(float)

df["Skor"] = df[["ADIL DAN AMANAH","SUKA MEMBANTU","NYAMBUNG KALAU NGOROL","ASYIK DAN MENARIK"]].dot(P)

print(df.head())
```

```python
#Ambil 4 mahasiswa terbaik
top4 = df.nlargest(4, "Skor")
print(top4)
```

```python
#Hitung matriks kovarian dan korelasi
cov_matrix = df[["ADIL DAN AMANAH","SUKA MEMBANTU","NYAMBUNG KALAU NGOROL","ASYIK DAN MENARIK"]].cov()
corr_matrix = df[["ADIL DAN AMANAH","SUKA MEMBANTU","NYAMBUNG KALAU NGOROL","ASYIK DAN MENARIK"]].corr()
print(cov_matrix)
print(corr_matrix)
```

```python
#Cosine similarity antar mahasiswa
from numpy.linalg import norm
import numpy as np

X = df[["ADIL DAN AMANAH","SUKA MEMBANTU","NYAMBUNG KALAU NGOROL","ASYIK DAN MENARIK"]].values
cosine_sim = (X @ X.T) / (norm(X, axis=1)[:, None] * norm(X, axis=1)[None, :])

cosine_df = pd.DataFrame(cosine_sim, index=df["NAMA"], columns=df["NAMA"])
print(cosine_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
#Data numerik
X = df[["ADIL DAN AMANAH","SUKA MEMBANTU","NYAMBUNG KALAU NGOROL","ASYIK DAN MENARIK"]].values

#PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
df["PC1"], df["PC2"] = X_pca[:,0], X_pca[:,1]

plt.figure(figsize=(8,6))
plt.scatter(df["PC1"], df["PC2"], c="purple", alpha=0.6)
for i, nama in enumerate(df["NAMA"]):
    plt.text(df["PC1"][i]+0.01, df["PC2"][i]+0.01, nama, fontsize=7)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Visualisasi PCA Mahasiswa")
plt.show()
```

Penjelasan Hasil PCA Mahasiswa

- PCA (Principal Component Analysis) digunakan untuk merangkum banyak variabel mahasiswa menjadi komponen utama sehingga bisa divisualisasikan lebih sederhana.

- PC1 (sumbu X) menunjukkan faktor utama yang paling membedakan mahasiswa, sedangkan PC2 (sumbu Y) menunjukkan faktor kedua.

- Setiap titik ungu pada grafik mewakili seorang mahasiswa, dan posisi titik ditentukan oleh hasil perhitungan PCA.

- Kedekatan antar titik menandakan kemiripan data mahasiswa, sementara titik yang berjauhan menunjukkan perbedaan signifikan.

- Mayoritas mahasiswa terkumpul di tengah → berarti karakteristik data mereka mirip/homogen.

- Beberapa mahasiswa terpencar (outlier) seperti di kanan atas atau bawah → artinya mereka punya karakteristik berbeda dari kelompok besar.

- Manfaat utama visualisasi ini adalah memudahkan dalam melihat pola, kelompok, dan perbedaan mahasiswa secara keseluruhan.

```python
#Clustering (K-Means)
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(df["PC1"], df["PC2"], c=df["Cluster"], cmap="viridis", alpha=0.7)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clustering Mahasiswa (K-Means, 3 cluster)")
plt.colorbar(label="Cluster")
plt.show()
```

Grafik di atas menampilkan hasil clustering mahasiswa menggunakan algoritma K-Means dengan 3 cluster. Data mahasiswa sebelumnya telah direduksi dimensinya dengan PCA sehingga bisa divisualisasikan ke dalam dua sumbu, yaitu PC1 dan PC2. Setiap titik pada grafik merepresentasikan seorang mahasiswa, dan warna titik menunjukkan kelompok atau cluster tempat mahasiswa tersebut tergabung. Dari hasil ini terlihat bahwa mahasiswa terbagi menjadi tiga kelompok utama dengan karakteristik yang berbeda. Kelompok berwarna ungu cenderung terkonsentrasi di sisi kiri, kelompok kuning lebih menyebar di bagian tengah, sedangkan kelompok hijau kebanyakan berada di sisi kanan grafik. Hal ini menunjukkan adanya perbedaan pola antar mahasiswa berdasarkan variabel awal yang digunakan. Dengan visualisasi clustering ini, dapat disimpulkan bahwa meskipun sebagian mahasiswa memiliki kesamaan karakteristik sehingga masuk dalam satu cluster, ada pula yang cukup berbeda sehingga dipisahkan ke cluster lain.

```python
#Cosine similarity
from numpy.linalg import norm
import numpy as np

X = df[["ADIL DAN AMANAH","SUKA MEMBANTU","NYAMBUNG KALAU NGOROL","ASYIK DAN MENARIK"]].values
cosine_sim = (X @ X.T) / (norm(X, axis=1)[:, None] * norm(X, axis=1)[None, :])

plt.figure(figsize=(10,8))
sns.heatmap(cosine_sim, xticklabels=df["NAMA"], yticklabels=df["NAMA"], cmap="YlGnBu")
plt.title("Cosine Similarity antar Mahasiswa")
plt.show()
```

**Kesimpulan Heatmap Cosine Similarity Mahasiswa**

1. Tingkat kemiripan tinggi secara umum

- Semua mahasiswa memiliki nilai cosine similarity di atas 0.90, artinya secara keseluruhan mereka punya karakteristik yang mirip dalam kriteria yang diukur (adil dan amanah, suka membantu, nyambung ngobrol, asyik dan menarik).

2. Diagonal = 1.00

- Nilai ini muncul pada perbandingan mahasiswa dengan dirinya sendiri, yang menandakan kesesuaian sempurna (identik).

3. Pasangan dengan kemiripan sangat tinggi (≈0.98–0.99)

- Beberapa mahasiswa terlihat sangat mirip satu sama lain karena warna kotaknya lebih terang (kuning–hijau muda).
Contoh: Rizky Faozi Mahendra ↔ Kiky Nurul Maqfiroh, serta Yumna Rachmawati ↔ Najwa Sofya.

4. Pasangan dengan kemiripan lebih rendah (≈0.90–0.93)

- Ditandai dengan warna biru tua. Contoh: Amelia Rizqyna Putri ↔ Madina Hedy Alifanti, yang artinya mereka punya perbedaan lebih jelas dibanding mahasiswa lain.

5. Adanya pola kelompok (cluster alami)

- Terlihat blok warna terang pada area tertentu, yang menunjukkan beberapa mahasiswa membentuk kelompok dengan kemiripan tinggi.

- Sementara mahasiswa dengan warna lebih gelap terhadap banyak orang lain bisa dianggap agak berbeda atau unik dalam karakteristiknya.
