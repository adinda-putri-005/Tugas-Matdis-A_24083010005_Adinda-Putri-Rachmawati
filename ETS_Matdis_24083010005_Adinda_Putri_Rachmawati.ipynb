# tabulasi konsonan-vokal per kata

```python
import pandas as pd
from string import ascii_lowercase

# Daftar huruf vokal & konsonan
vowels = set("aeiou")
consonants = set("bcdfghjklmnpqrstvwxyz")

# Fungsi bantu untuk menghitung jumlah huruf

def hitung_konsonan_vokal(kata):
    kata = kata.lower()
    v = sum(1 for c in kata if c in vowels)
    k = sum(1 for c in kata if c in consonants)
    return k, v

def proses_bagian(teks, nama_bagian):
    kata_list = re.findall(r"[a-zA-Z']+", teks)
    data = []
    for kata in kata_list:
        k, v = hitung_konsonan_vokal(kata)
        data.append({
            "Bagian": nama_bagian,
            "Kata": kata.lower(),
            "Konsonan": k,
            "Vokal": v
        })
    return data

# Proses tiap bagian jadi satu DataFrame

semua_data = []
semua_data.extend(proses_bagian(abstrak, "Abstrak"))
semua_data.extend(proses_bagian(pendahuluan, "Pendahuluan"))
semua_data.extend(proses_bagian(kesimpulan, "Kesimpulan"))

df = pd.DataFrame(semua_data)

print("=== DATA CONTOH (10 BARIS PERTAMA) ===\n")
print(df.head(10))

# Tabulasi per bagian (dipisah)


for bagian in df["Bagian"].unique():
    subset = df[df["Bagian"] == bagian]
    tabulasi = pd.crosstab(subset["Vokal"], subset["Konsonan"])

    print(f"\n\n=== TABULASI KOMBINASI KONSONAN–VOKAL: {bagian.upper()} ===\n")
    print(tabulasi)

    # Analisis ringkas per bagian

    if not tabulasi.empty:
        max_combo = tabulasi.stack().idxmax()
        max_value = tabulasi.stack().max()

        print("Bahasan Ringkas:")
        print(f"- Pada bagian {bagian.lower()}, kombinasi paling sering muncul adalah "
              f"{max_combo[1]} konsonan dan {max_combo[0]} vokal.")
        print(f"- Kombinasi tersebut muncul sebanyak {max_value} kali.")
        print("- Pola ini menunjukkan struktur kata paling sering digunakan pada bagian ini.")
        print("- Kata dengan banyak konsonan biasanya kata teknis, sedangkan sedikit konsonan lebih umum atau pendek.")
    else:
        print("\n(Tidak ada data untuk bagian ini)")
```

```python
def hitung_konsonan_vokal(kata):
    kata = kata.lower()
    v = sum(1 for c in kata if c in vowels)
    k = sum(1 for c in kata if c in consonants)
    return k, v

def proses_bagian(teks, nama_bagian):
    # Ambil semua kata dari teks
    kata_list = re.findall(r"[a-zA-Z']+", teks)
    data = []
    for kata in kata_list:
        k, v = hitung_konsonan_vokal(kata)
        data.append({
            "Bagian": nama_bagian,
            "Konsonan": k,
            "Vokal": v
        })
    return pd.DataFrame(data)

# ============================================================
# 4️⃣ Buat Pivot Table per Bagian
# ============================================================

for bagian, teks in [("Abstrak", abstrak), ("Pendahuluan", pendahuluan), ("Kesimpulan", kesimpulan)]:
    df = proses_bagian(teks, bagian)

    # Pivot: sumbu X = konsonan, sumbu Y = vokal
    pivot = pd.pivot_table(
        df,
        index="Vokal",        # sumbu Y
        columns="Konsonan",   # sumbu X
        aggfunc="size",       # hitung banyaknya kombinasi
        fill_value=0
    )

    # ============================================================
    # 5️⃣ Cetak Pivot dengan label yang jelas
    # ============================================================

    print(f"\n{'='*70}")
    print(f"PIVOT TABLE KONSONAN–VOKAL ({bagian.upper()})")
    print(f"{'='*70}\n")
    print(" Keterangan:")
    print("- Kolom (X-axis) : Jumlah huruf KONSONAN")
    print("- Baris (Y-axis) : Jumlah huruf VOKAL")
    print("- Isi tabel      : Banyaknya kata dengan kombinasi tersebut\n")

    print(pivot)

    # ============================================================
    # 6️⃣ Analisis Ringkas
    # ============================================================

    if not pivot.empty:
        max_combo = pivot.stack().idxmax()
        max_value = pivot.stack().max()
        print(" Bahasan Ringkas:")
        print(f"- Kombinasi paling sering di bagian {bagian.lower()} adalah:")
        print(f"  ➤ {max_combo[1]} konsonan dan {max_combo[0]} vokal ({max_value} kata).")
        print("- Semakin besar nilai pada kolom kanan (X-axis), berarti banyak kata panjang.")
        print("- Semakin besar nilai pada baris bawah (Y-axis), berarti banyak kata dengan banyak vokal.")
    else:
        print("\n(Tidak ada data ditemukan untuk bagian ini)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import re
import pandas as pd

# Definisi huruf vokal dan konsonan
vowels = "aiueo" # Changed from vokal
consonants = "bcdfghjklmnpqrstvwxyz" # Changed from konsonan

# Combine all words from all sections
all_text = abstrak + pendahuluan + kesimpulan
daftar_kata = re.findall(r"[a-zA-Z']+", all_text)


# Buat dictionary untuk frekuensi dan jumlah kata
freq_kombinasi = {k: {v: 0 for v in vowels} for k in consonants} # Changed from vokal to vowels and konsonan to consonants
kata_kombinasi = {k: {v: set() for v in vowels} for k in consonants} # Changed from vokal to vowels and konsonan to consonants

# Ekstraksi kombinasi dari setiap kata
for kata in daftar_kata:
    kata = kata.lower()
    for i in range(len(kata) - 1):
        c, v = kata[i], kata[i + 1]
        if c in consonants and v in vowels: # Changed from konsonan to consonants and vokal to vowels
            freq_kombinasi[c][v] += 1
            kata_kombinasi[c][v].add(kata)

# Konversi ke DataFrame
df_freq = pd.DataFrame(freq_kombinasi).T[list(vowels)] # Changed from vokal to vowels and added list()
df_kata = pd.DataFrame({k: {v: len(kata_kombinasi[k][v]) for v in vowels} for k in consonants}).T[list(vowels)] # Changed from vokal to vowels and added list()

# ==============================
# 🔹 HEATMAP + TABEL FREKUENSI
# ==============================
plt.figure(figsize=(8, 6))
sns.heatmap(df_freq, annot=True, cmap="YlOrRd", fmt="d")
plt.title("Frekuensi Kombinasi Konsonan–Vokal")
plt.xlabel("Vokal")
plt.ylabel("Konsonan")
plt.show()

print("=== 📊 TABEL FREKUENSI KONSONAN–VOKAL ===")
display(df_freq.fillna(0))

# ==============================
# 🔹 HEATMAP + TABEL JUMLAH KATA
# ==============================
plt.figure(figsize=(8, 6))
sns.heatmap(df_kata, annot=True, cmap="BuPu", fmt="d")
plt.title("Jumlah Kata yang Mengandung Kombinasi Konsonan–Vokal")
plt.xlabel("Vokal")
plt.ylabel("Konsonan")
plt.show()

print("=== 📘 TABEL JUMLAH KATA PER KOMBINASI ===")
display(df_kata.fillna(0))
```

**Penjelasan singkat frekuensi kombinasi konsonan-vokal:**

- Sumbu vertikal (y) menunjukkan konsonan, sedangkan sumbu horizontal (x) menunjukkan vokal (a, i, u, e, o).

- Warna menunjukkan tingkat frekuensi: semakin merah tua, semakin sering kombinasi tersebut muncul.

- Dari peta terlihat bahwa kombinasi seperti “t-a”, “p-i”, dan “s-a” memiliki frekuensi tertinggi, ditandai warna merah gelap.

- Sementara kombinasi lain seperti “q-e” atau “z-u” jarang atau bahkan tidak muncul (warna kuning muda).

- Kesimpulannya, ada pola tertentu di mana konsonan tertentu (misal t, p, s) lebih sering digunakan bersama vokal tertentu (terutama a dan i).

**Penjelasan singkat Jumlah kata yang mengandung Kombinasi Konsonan-Vokal**
- Sumbu vertikal (kiri) = konsonan, dan sumbu horizontal (bawah) = vokal (a, i, u, e, o).

- Warna ungu gelap menunjukkan kombinasi yang sering muncul dalam banyak kata, sedangkan warna biru muda/pucat menunjukkan kombinasi yang jarang muncul.

- Kombinasi seperti “s-a”, “p-i”, dan “t-a” tampak memiliki jumlah kata terbanyak, terlihat dari warna paling gelap.

- Sebaliknya, kombinasi dengan konsonan seperti “q”, “x”, atau “z” relatif jarang digunakan.

- Kesimpulannya, pola ini menunjukkan bahwa dalam kumpulan data, konsonan umum (t, p, s, n) lebih sering muncul dalam kata yang mengandung vokal a atau i, sedangkan kombinasi dengan konsonan jarang seperti q atau x hampir tidak muncul sama sekali.
