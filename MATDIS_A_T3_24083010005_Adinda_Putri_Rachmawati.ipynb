```python
#nomor 1
def cek_pembagian(n):
    if n == 1:
        return "Tidak terdefinisi (n-1 = 0)"
    return (n**2 - n) % (n-1) == 0  # True kalau habis dibagi, False kalau tidak

# Coba untuk beberapa nilai n
for n in range(2, 11):  # mulai dari 2 biar gak kena pembagian 0
    print(f"n={n}, hasil={cek_pembagian(n)}")
```

```python
#nomor 2
def cari_n_maks_sebelum_2(limit=100):
    jumlah = 0
    n_maks = 0
    for n in range(1, limit+1):
        jumlah += 1/n
        print(f"n={n}, Hn={jumlah}")  # tampilkan proses
        if jumlah < 2:
            n_maks = n
        else:
            break
    return n_maks

n_maks = cari_n_maks_sebelum_2()
print(f"\nJadi n maksimal sebelum hasilnya 2 adalah {n_maks}")
```

```python
def cek_tepat_2(limit=1000):
    H = 0.0
    n_maks = 0
    for n in range(1, limit+1):
        H += 1/n
        if H == 2:        # cek apakah jumlah tepat 2
            return True, n
        if H < 2:         # simpan n terakhir sebelum lewat 2
            n_maks = n
    return False, n_maks

# cek sampai 1000 suku
ada, n = cek_tepat_2(1000)

if ada:
    print(f"True → ditemukan n dengan Hn = 2, yaitu n={n}")
else:
    print(f"False → tidak ada Hn = 2, n maksimal sebelum melewati 2 adalah {n}")
```

```python
import matplotlib.pyplot as plt

# Hitung deret harmonik untuk beberapa n
n_values = list(range(1, 11))  # ganti 11 jadi angka lebih besar kalau mau sampai n=50 atau n=100
H_values = [sum(1/i for i in range(1, n+1)) for n in n_values]

# Plot garis deret harmonik
plt.figure(figsize=(8,5))
plt.plot(n_values, H_values, marker='o', label='Hn (Deret Harmonik)')
plt.axhline(y=2, color='r', linestyle='--', label='y = 2')  # garis pembanding

# Tambahkan highlight untuk n=3 dan n=4
plt.scatter([3,4], [H_values[2], H_values[3]], color='orange', zorder=5)
plt.text(3, H_values[2]-0.1, f"H3={H_values[2]:.3f}", ha='center', color='orange')
plt.text(4, H_values[3]+0.05, f"H4={H_values[3]:.3f}", ha='center', color='orange')

# Judul dan label
plt.title("Plot Deret Harmonik Hn dibandingkan dengan 2")
plt.xlabel("n")
plt.ylabel("Hn")
plt.legend()
plt.grid(True)
plt.show()
```


### **Penjelasan Grafik Deret Harmonik (Hn) dibandingkan dengan y = 2**

Grafik memperlihatkan nilai deret harmonik parsial:

$$
H_n = 1 + \frac{1}{2} + \frac{1}{3} + \dots + \frac{1}{n}
$$

yang diplot terhadap variabel $n$. Garis horizontal merah putus-putus menunjukkan batas $y = 2$.

* Pada titik $n = 1$, $H_1 = 1$.
* Pada $n = 2$, $H_2 = 1.5$.
* Pada $n = 3$, $H_3 \approx 1.833$, nilai ini **masih di bawah 2**.
* Pada $n = 4$, $H_4 \approx 2.083$, nilai ini **melewati batas 2**.

pada gambar diatas terlihat jelas pada grafik bahwa deret harmonik $H_n$ bertambah secara monoton (selalu naik) seiring bertambahnya $n$, namun pertambahannya semakin kecil.

---

### **Kesimpulan dari Grafik**

* Garis merah $y = 2$ berfungsi sebagai pembanding.
* Titik $H_3$ adalah nilai terakhir yang masih berada di bawah garis merah.
* Titik $H_4$ sudah melampaui garis merah.

Jadi, **nilai maksimum $n$ yang masih menghasilkan $H_n < 2$ adalah $n = 3$**.
